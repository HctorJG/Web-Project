```python
url = 'https://www.goodreads.com/list/show/24366.Best_machine_learning_books'
response = requests.get(url)
html = response.content
soup = BeautifulSoup(html, 'lxml')
titulo = soup.select('a[class="bookTitle"]')
libro_titulo = [element.text for element in titulo]
clean_libro = [text.strip() for text in libro_titulo]
clean_libro
```

```python
url = 'https://www.goodreads.com/list/show/24366.Best_machine_learning_books'
response = requests.get(url)
html = response.content
soup = BeautifulSoup(html, 'lxml')
titulo = soup.select('a[class="authorName"]')
autor_titulo = [element.text for element in titulo]
clean_autor = [text.strip() for text in autor_titulo]
clean_autor
```

```python
url = 'https://www.goodreads.com/list/show/24366.Best_machine_learning_books'
response = requests.get(url)
html = response.content
soup = BeautifulSoup(html, 'lxml')
titulo = soup.select('[class="minirating"]')
rating_titulo = [element.text for element in titulo]
clean_rating = [text.strip().replace('it was amazing ', '').replace('liked it ','')
                .replace('really ', '').replace('it was ok ', '')
                .replace(' avg rating — ', ' de ').replace('ratings', 'votos') for text in rating_titulo]
clean_rating
```

```python
len (clean_libro), len(clean_autor), len(clean_rating)
```

```python
df = pd.DataFrame(zip(clean_libro, clean_autor, clean_rating),
                  columns=['Libros Machine Learning', 'Autor', 'Calificacion escala 1 al 5'])
```
